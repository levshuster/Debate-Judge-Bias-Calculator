```python
import requests
from bs4 import BeautifulSoup

def get_formats(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    first_dropdown = soup.find('select')
    if first_dropdown:
        return [
            (option.get('value'), option.text.strip())
            for option in first_dropdown.find_all('option')[1:]
        ]
    else:
        return [None, "No Options Found"]

url = "https://www.tabroom.com/index/tourn/postings/index.mhtml?tourn_id=26620"
get_formats(url)
```

```python
import requests
from bs4 import BeautifulSoup

# URL to send the GET request to
url = "https://www.tabroom.com/index/tourn/postings/index.mhtml?tourn_id=26620"

# Send the GET request
response = requests.get(url)

# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find the first dropdown (select element)
first_dropdown = soup.find('select')

# Check if the dropdown exists
if first_dropdown:
    # Find all options within the dropdown
    options = first_dropdown.find_all('option')

    # Loop through the options and print the value and the shown text
    for option in options:
        value = option.get('value')
        text = option.text.strip()  # Strip to remove any unnecessary whitespace
        print(f"Option Value: {value}, Shown Text: {text}")
else:
    print("No dropdown found.")
```
